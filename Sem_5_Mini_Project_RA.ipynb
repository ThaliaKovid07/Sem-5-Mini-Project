### Now let try KNN classifier on Weather Dataset

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = weather_df.iloc[:,1:18]
y = weather_df.iloc[:,19]
```

```python
plt.figure(figsize=(25,7))
sns.countplot(x='Cause_Weather',hue='W_fatility',data=weather_df,palette='Set1')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=9,p=2,metric='euclidean')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test,y_pred)
print(conf_matrix)
print(f1_score(y_test,y_pred))
```

Accuracy of KNN Classifier

```python
print(accuracy_score(y_test,y_pred))
```

```python
conditions = [
		'Total Accidents National Highways under NHAI' == roadtype_df[['Total Accidents National Highways under NHAI','Total Accidents National Highways under State PWD','Total AccidentsNational Highways under others','Total Accidents State Highways','Total Accidents Others Road']].idxmax(axis=1),
		'Total Accidents National Highways under State PWD' == roadtype_df[['Total Accidents National Highways under NHAI','Total Accidents National Highways under State PWD','Total AccidentsNational Highways under others','Total Accidents State Highways','Total Accidents Others Road']].idxmax(axis=1),
		'Total AccidentsNational Highways under others' == roadtype_df[['Total Accidents National Highways under NHAI','Total Accidents National Highways under State PWD','Total AccidentsNational Highways under others','Total Accidents State Highways','Total Accidents Others Road']].idxmax(axis=1),
		'Total Accidents State Highways' == roadtype_df[['Total Accidents National Highways under NHAI','Total Accidents National Highways under State PWD','Total AccidentsNational Highways under others','Total Accidents State Highways','Total Accidents Others Road']].idxmax(axis=1),
		'Total Accidents Others Road' == roadtype_df[['Total Accidents National Highways under NHAI','Total Accidents National Highways under State PWD','Total AccidentsNational Highways under others','Total Accidents State Highways','Total Accidents Others Road']].idxmax(axis=1)
            ]

values =  ['NH under NHAI', 'NH under State PWD','NH under other','State Highway','Other Road']
roadtype_df['Cause_roadtype'] = np.select(conditions, values)
roadtype_df.head(1)
```

Note:


*   mn is mean
*   NH is National Highway


*   St is State
*   Sth is State Highway

*   OtH is otherroad


```python
mn_NH_under_NHAI=roadtype_df['Total Deaths National Highways under NHAI'].mean()
print(mn_NH_under_NHAI)
mn_NH_under_St_pwd=roadtype_df['Total Deaths National Highways under State PWD'].mean()
print(mn_NH_under_St_pwd)
mn_NH_under_other=roadtype_df['Total Deaths National Highways under others'].mean()
print(mn_NH_under_other)
mn_StH=roadtype_df['Total Deaths State Highways'].mean()
print(mn_StH)
mn_OtH=roadtype_df['Total Deaths Others Road'].mean()
print(mn_OtH)
```

```python
for ind in roadtype_df.index:
  if (roadtype_df['Cause_roadtype'][ind]=='NH under NHAI'):
    if (roadtype_df['Total Deaths National Highways under NHAI'][ind]>mn_NH_under_NHAI):
      roadtype_df['roadtype_fatility'][ind]='High'
    else:
      roadtype_df['roadtype_fatility'][ind]='Low'
  elif (roadtype_df['Cause_roadtype'][ind]=='NH under State PWD'):
    if (roadtype_df['Total Deaths National Highways under State PWD'][ind]>mn_NH_under_St_pwd):
      roadtype_df['roadtype_fatility'][ind]='High'
    else:
      roadtype_df['roadtype_fatility'][ind]='Low'
  elif (roadtype_df['Cause_roadtype'][ind]=='NH under other'):
    if (roadtype_df['Total Deaths National Highways under others'][ind]>mn_NH_under_other):
      roadtype_df['roadtype_fatility'][ind]='High'
    else:
      roadtype_df['roadtype_fatility'][ind]='Low'
  elif (roadtype_df['Cause_roadtype'][ind]=='State Highway'):
    if (roadtype_df['Total Deaths State Highways'][ind]>mn_StH):
      roadtype_df['roadtype_fatility'][ind]='High'
    else:
      roadtype_df['roadtype_fatility'][ind]='Low'
  elif (roadtype_df['Cause_roadtype'][ind]=='Other Road'):
    if (roadtype_df['Total Deaths Others Road'][ind]>mn_OtH):
      roadtype_df['roadtype_fatility'][ind]='High'
    elif (roadtype_df['Total Deaths Others Road'][ind]<mn_OtH):
      roadtype_df['roadtype_fatility'][ind]='Low'
```
